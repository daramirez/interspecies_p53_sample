```r
# Defines function that plots a fixed species against all other nine species log2FoldChange.

pairwiseDEGscatterplot <- function(library, fixedSpecies) {
    
    start.time <- Sys.time()

    # Removes the fixed species from the species list.
    speciesListAll <- c("Human","Chimp","Bonobo","Gorilla","Orangutan",
                        "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")
    speciesList <- speciesListAll[!(speciesListAll %in% fixedSpecies)]

    # Creates color scheme data frame that matches species with their color.
    colorScheme <- data.frame(matrix(ncol = 1, nrow = length(speciesListAll)))
    colnames(colorScheme) <- "colorCode"
    rownames(colorScheme) <- speciesListAll
    colorScheme$colorCode <- append("black",viridis(9))

    for(otherSpecies in speciesList) {

        # Merges the DEGs per species pair.
        p53_names_pair <- eval(as.symbol(paste0("names_sig_",library,"_", fixedSpecies)))
        other <- eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies)))[!(eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies))) %in% p53_names_pair)]
        lengthOther <- length(eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies))))
        checkOther <- sum(eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies))) %in% eval(as.symbol(paste0("names_sig_",library,"_", fixedSpecies))))
        if(length(other) == 0 & lengthOther == checkOther) { p53_names_pair <- p53_names_pair }
        if(length(other) == 1) {if(is.na(other)) { p53_names_pair <- p53_names_pair } else { p53_names_pair <- c(p53_names_pair, other) }}
        if(length(other) > 1) { p53_names_pair <- c(p53_names_pair, other) }
        
        # Initializes data frame that will contain the species pair categorization of DEGs.
        colNames <- c("speciesA_fc","speciesB_fc","speciesA_sig","speciesB_sig","both_sig","label_sig","species_sig")
        speciespairSignificance <- data.frame(matrix(FALSE, ncol = length(colNames), nrow = length(p53_names_pair)))
        colnames(speciespairSignificance) <- colNames
        rownames(speciespairSignificance) <- p53_names_pair

        # Fills in data frame with log2FoldChange and per species binary significance category.

        for(gene in p53_names_pair) {
            speciespairSignificance[gene,"speciesA_fc"] <- eval(as.symbol(paste0("results_all_",library,"_", fixedSpecies)))[gene,"log2FoldChange"]
            speciespairSignificance[gene,"speciesB_fc"] <- eval(as.symbol(paste0("results_all_",library,"_", otherSpecies)))[gene,"log2FoldChange"]
            if(gene %in% eval(as.symbol(paste0("names_sig_",library,"_", fixedSpecies)))) 
            { speciespairSignificance[gene,"speciesA_sig"] <- TRUE }
            if(gene %in% eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies)))) 
            { speciespairSignificance[gene,"speciesB_sig"] <- TRUE }
            if (speciespairSignificance[gene,"speciesA_sig"] & speciespairSignificance[gene,"speciesB_sig"])
            { speciespairSignificance[gene,"both_sig"] <- TRUE
            speciespairSignificance[gene,"label_sig"] <- "both"}
            if (isTRUE(speciespairSignificance[gene,"speciesA_sig"]) & 
                !isTRUE(speciespairSignificance[gene,"speciesB_sig"]))
            { speciespairSignificance[gene,"label_sig"] <- "speciesA" }
            if (!isTRUE(speciespairSignificance[gene,"speciesA_sig"]) & 
                isTRUE(speciespairSignificance[gene,"speciesB_sig"]))
            { speciespairSignificance[gene,"label_sig"] <- "speciesB" }
            if (eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 1 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 2 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 3) {
                speciespairSignificance[gene,"species_sig"] <- "1_2_3" }
            if (eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 4 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 5 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 6) {
                speciespairSignificance[gene,"species_sig"] <- "4_5_6" }
            if (eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 7 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 8 | 
                eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 9 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 10) {
                speciespairSignificance[gene,"species_sig"] <- "7_8_9_10" } }

        # Removes entries that have NA in either species log2FoldChange.
        speciespairSignificance <- speciespairSignificance[complete.cases(speciespairSignificance$speciesA_fc),]
        speciespairSignificance <- speciespairSignificance[complete.cases(speciespairSignificance$speciesB_fc),]

        # Finds the maximum value to adjust the length of both axes.
        maxAxis <- max(c(speciespairSignificance$speciesA_fc,speciespairSignificance$speciesB_fc)) + 
                   (max(c(speciespairSignificance$speciesA_fc,speciespairSignificance$speciesB_fc))*0.1)
        minAxis <- min(c(speciespairSignificance$speciesA_fc,speciespairSignificance$speciesB_fc)) + 
                   (min(c(speciespairSignificance$speciesA_fc,speciespairSignificance$speciesB_fc))*0.1)

        # Stores each plot.
        fcPlot <- ggplot(data = speciespairSignificance, aes(x = speciesB_fc, y = speciesA_fc, color = label_sig)) +
          geom_point(shape = 19, size = 3) +
          xlim(minAxis, maxAxis) +
          ylim(minAxis, maxAxis) +
          theme_minimal(base_size = 20) +
          geom_hline(yintercept = 0, color = "black", size = 1) +
          geom_vline(xintercept = 0, color = "black", size = 1) +
          labs(title = paste0(fixedSpecies," vs ",otherSpecies,"\nn = ", dim(speciespairSignificance)[1]),
               x = otherSpecies,
               y = fixedSpecies) +
          scale_color_manual("",
                             labels = c(paste0(fixedSpecies,"   n = ",as.character(table(speciespairSignificance$label_sig)["speciesA"])),
                                        paste0(otherSpecies,"   n = ",as.character(table(speciespairSignificance$label_sig)["speciesB"])),
                                        paste0("Both   n = ",as.character(table(speciespairSignificance$label_sig)["both"]))), 
                             breaks = c("speciesA","speciesB","both"),
                             values = c(colorScheme[fixedSpecies,],colorScheme[otherSpecies,],"darkgrey")) +
          theme(legend.position = c(1,0), legend.justification = c(1,0)) +
          theme(plot.title = element_text(hjust = 0.5),
                axis.line = element_blank())

        assign(paste0("plot_",fixedSpecies,"_",otherSpecies), fcPlot)
    }

    # Arranges all plots onto a single grid.
    options(repr.plot.width = 26, repr.plot.height = 13.5)
    suppressMessages(grid.arrange(eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[1]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[2]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[3]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[4]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[5]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[6]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[7]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[8]))), 
                 eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[9]))), 
                 ncol = 5, nrow = 2,
                 top = textGrob(paste0(library,"-seq  |  log2FC  |  ",fixedSpecies," vs others\n"), 
                                gp = gpar(fontsize = 35))))
    
    end.time <- Sys.time()
    time.taken <- end.time - start.time
    time.taken
}
```

```r
#########################################
             # FOR POSTER 
#########################################

start.time <- Sys.time()

library <- "PRO"
fixedSpecies <- "Human"

# Removes the fixed species from the species list.
speciesListAll <- c("Human","Chimp","Bonobo","Gorilla","Orangutan",
                    "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")
speciesList <- speciesListAll[!(speciesListAll %in% fixedSpecies)]

# Creates color scheme data frame that matches species with their color.
colorScheme <- data.frame(matrix(ncol = 1, nrow = length(speciesListAll)))
colnames(colorScheme) <- "colorCode"
rownames(colorScheme) <- speciesListAll
colorScheme$colorCode <- append("black",viridis(9))

for(otherSpecies in speciesList) {

    # Merges the DEGs per species pair.
    p53_names_pair <- eval(as.symbol(paste0("names_sig_",library,"_", fixedSpecies)))
    other <- eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies)))[!(eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies))) %in% p53_names_pair)]
    lengthOther <- length(eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies))))
    checkOther <- sum(eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies))) %in% eval(as.symbol(paste0("names_sig_",library,"_", fixedSpecies))))
    if(length(other) == 0 & lengthOther == checkOther) { p53_names_pair <- p53_names_pair }
    if(length(other) == 1) {if(is.na(other)) { p53_names_pair <- p53_names_pair } else { p53_names_pair <- c(p53_names_pair, other) }}
    if(length(other) > 1) { p53_names_pair <- c(p53_names_pair, other) }

    # Initializes data frame that will contain the species pair categorization of DEGs.
    colNames <- c("speciesA_fc","speciesB_fc","speciesA_sig","speciesB_sig","both_sig","label_sig","species_sig")
    speciespairSignificance <- data.frame(matrix(FALSE, ncol = length(colNames), nrow = length(p53_names_pair)))
    colnames(speciespairSignificance) <- colNames
    rownames(speciespairSignificance) <- p53_names_pair

    # Fills in data frame with log2FoldChange and per species binary significance category.

    for(gene in p53_names_pair) {
        speciespairSignificance[gene,"speciesA_fc"] <- eval(as.symbol(paste0("results_all_",library,"_", fixedSpecies)))[gene,"log2FoldChange"]
        speciespairSignificance[gene,"speciesB_fc"] <- eval(as.symbol(paste0("results_all_",library,"_", otherSpecies)))[gene,"log2FoldChange"]
        if(gene %in% eval(as.symbol(paste0("names_sig_",library,"_", fixedSpecies)))) 
        { speciespairSignificance[gene,"speciesA_sig"] <- TRUE }
        if(gene %in% eval(as.symbol(paste0("names_sig_",library,"_", otherSpecies)))) 
        { speciespairSignificance[gene,"speciesB_sig"] <- TRUE }
        if (speciespairSignificance[gene,"speciesA_sig"] & speciespairSignificance[gene,"speciesB_sig"])
        { speciespairSignificance[gene,"both_sig"] <- TRUE
        speciespairSignificance[gene,"label_sig"] <- "both"}
        if (isTRUE(speciespairSignificance[gene,"speciesA_sig"]) & 
            !isTRUE(speciespairSignificance[gene,"speciesB_sig"]))
        { speciespairSignificance[gene,"label_sig"] <- "speciesA" }
        if (!isTRUE(speciespairSignificance[gene,"speciesA_sig"]) & 
            isTRUE(speciespairSignificance[gene,"speciesB_sig"]))
        { speciespairSignificance[gene,"label_sig"] <- "speciesB" }
        if (eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 1 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 2 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 3) {
            speciespairSignificance[gene,"species_sig"] <- "1_2_3" }
        if (eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 4 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 5 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 6) {
            speciespairSignificance[gene,"species_sig"] <- "4_5_6" }
        if (eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 7 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 8 | 
            eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 9 | eval(as.symbol(paste0("genes_sig_all_",library)))[gene,"Number"] == 10) {
            speciespairSignificance[gene,"species_sig"] <- "7_8_9_10" } }

    # Removes entries that have NA in either species log2FoldChange.
    speciespairSignificance <- speciespairSignificance[complete.cases(speciespairSignificance$speciesA_fc),]
    speciespairSignificance <- speciespairSignificance[complete.cases(speciespairSignificance$speciesB_fc),]

    # Stores each plot.
    fcPlot <- ggplot(data = speciespairSignificance, aes(x = speciesB_fc, y = speciesA_fc, color = label_sig)) +
      geom_point(shape = 19, size = 6) +
#       xlim(-3, 5) +
#       ylim(-3, 5) +
      xlim(-1.5, 1.5) +
      ylim(-1.5, 1.5) +
      theme_minimal(base_size = 30) +
      geom_hline(yintercept = 0, color = "black", size = 1) +
      geom_vline(xintercept = 0, color = "black", size = 1) +
      labs(title = paste0("n = ", dim(speciespairSignificance)[1]),
           x = otherSpecies,
           y = fixedSpecies) +
      scale_color_manual("",
                         labels = c(paste0(fixedSpecies,"   n = ",as.character(table(speciespairSignificance$label_sig)["speciesA"])),
                                    paste0(otherSpecies,"   n = ",as.character(table(speciespairSignificance$label_sig)["speciesB"])),
                                    paste0("Both   n = ",as.character(table(speciespairSignificance$label_sig)["both"]))), 
                         breaks = c("speciesA","speciesB","both"),
                         values = c(colorScheme[fixedSpecies,],colorScheme[otherSpecies,],"darkgrey")) +
      theme(legend.title = element_blank(),
            legend.position = c(1,0), 
            legend.justification = c(1,0),
            legend.direction = "vertical",
            legend.box.background = element_rect(color = "black", size = 1),
            axis.title.y = element_blank(),
            axis.title.x = element_blank()) +
      theme(plot.title = element_text(hjust = 0.5),
            axis.line = element_blank())

    assign(paste0("plot_",fixedSpecies,"_",otherSpecies), fcPlot)
}

# Arranges all plots onto a single grid.
options(repr.plot.width = 50, repr.plot.height = 6)
suppressMessages(grid.arrange(eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[1]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[2]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[3]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[4]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[5]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[6]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[7]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[8]))), 
             eval(as.symbol(paste0("plot_",fixedSpecies,"_",speciesList[9]))), 
             nrow = 1))

end.time <- Sys.time()
time.taken <- end.time - start.time
time.taken
```
