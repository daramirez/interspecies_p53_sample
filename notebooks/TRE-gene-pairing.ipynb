# Assigning TREs with their putative target genes using Hi-C

## Downloading public Hi-C datasets

I downloaded the raw fastq files from Hi-C datasets that were obtained from several non-treated primate LCLs.

## Using already processed Hi-C loops.

```
wget https://ftp.ncbi.nlm.nih.gov/geo/series/GSE63nnn/GSE63525/suppl/GSE63525_GM12878_combined_intrachromosomal_contact_matrices.tar.gz
wget https://ftp.ncbi.nlm.nih.gov/geo/series/GSE63nnn/GSE63525/suppl/GSE63525_GM12878_primary+replicate_HiCCUPS_looplist.txt.gz
wget https://ftp.ncbi.nlm.nih.gov/geo/series/GSE63nnn/GSE63525/suppl/GSE63525_GM12878_primary_README.rtf
wget https://ftp.ncbi.nlm.nih.gov/geo/series/GSE63nnn/GSE63525/suppl/GSE63525_CH12_combined_30.hic

# Editing HiCCUPS_looplist.txt onto a longrange file for visualization.
# Removes the first header line.
tail -n +2 GSE63525_GM12878_primary+replicate_HiCCUPS_looplist.txt > test0.bed
# Keeps only the columns with the coordinates, and the score.
awk -v OFS='\t' '{print "chr"$1, $2, $3, "chr"$4, $5, $6, $8}' test0.bed > test1.bed
# Keeps only entries with a score greater than 100.
awk -v OFS='\t' '{if ($7 > 100) print $0}' test1.bed > test2.bed
# Edits entries so that they resemble the 'longrange' format.
awk -v OFS='\t' '{ print $1, $2, $3, $4":"$5"-"$6","$7"\n"$4, $5, $6, $1":"$2"-"$3","$7}' test2.bed > test3.bed
# Sorts entries.
sort -k1,1 -k2,2n test3.bed > test4.bed
#Compresses with bgzip.
module load samtools/1.10
bgzip test4.bed
#Creates index with samtools tabix.
tabix -p bed test4.bed.gz
```
