````r
####################################
####################################
         # Running DESeq2
####################################
####################################

cellline <- "HCT116"
species <- "Human"

contrast <- c("treatment", "Nutlin", "DMSO")
alphaValue <- 0.5

baseDir <- paste0("/scratch/Users/dara6367/PRO-seq_interspecies-nutlin/Bidirectional-Flow/",species,"/")
countsDir <- paste0(baseDir,"featureCounts/")

####################################
           # Tfit bidirs
####################################

# # Reads in count data and their conditions table.
# load(paste0(countsDir,"PRO-",cellline,".tfit.featureCounts.RData"))
# conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-",cellline,".csv")
# conditionsTable_tfit <- read.csv(conditionsTableFile, header = TRUE)

# # Assigns the same ID to both the conditionTables row names and the counts column names.
# colnames(counts$counts) <- conditionsTable_tfit$sample
# rownames(conditionsTable_tfit) <- conditionsTable_tfit$sample
# counts_tfit <- counts$counts

# # Creates a DESeq2Dataset object and loads in counts.
# dds_tfit <- DESeqDataSetFromMatrix(countData = counts_tfit, 
#                               colData = conditionsTable_tfit,
#                               design = ~ treatment)

# # Runs DESeq2 using the counts matrix.
# dds_tfit <- DESeq(dds_tfit)

# # 'Shrink' results using 'normal' mode.
# resultsUnshrunken_tfit <- results(dds_tfit, 
#                              contrast = contrast, 
#                              alpha = alphaValue)

# results_tfit <- lfcShrink(dds_tfit, 
#                      contrast = contrast, 
#                      res = resultsUnshrunken_tfit, 
#                      type = "normal")

# results_tfit_filt <- results_tfit[complete.cases(results_tfit$padj),]
# results_tfit_filt <- results_tfit_filt$padj < alphaValue
# tfit_n <- as.character(table(results_tfit_filt)["TRUE"])

####################################
           # dREG bidirs
####################################

# # Reads in count data and their conditions table.
# load(paste0(countsDir,"PRO-",cellline,".dreg.featureCounts.RData"))
# conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-",cellline,".csv")
# conditionsTable_dreg <- read.csv(conditionsTableFile, header = TRUE)

# # Assigns the same ID to both the conditionTables row names and the counts column names.
# colnames(counts$counts) <- conditionsTable_dreg$sample
# rownames(conditionsTable_dreg) <- conditionsTable_dreg$sample
# counts_dreg <- counts$counts

# # Creates a DESeq2Dataset object and loads in counts.
# dds_dreg <- DESeqDataSetFromMatrix(countData = counts_dreg, 
#                               colData = conditionsTable_dreg,
#                               design = ~ treatment)

# # Runs DESeq2 using the counts matrix.
# dds_dreg <- DESeq(dds_dreg)

# # 'Shrink' results using 'normal' mode.
# resultsUnshrunken_dreg <- results(dds_dreg, 
#                              contrast = contrast, 
#                              alpha = alphaValue)

# results_dreg <- lfcShrink(dds_dreg, 
#                      contrast = contrast, 
#                      res = resultsUnshrunken_dreg, 
#                      type = "normal")

# results_dreg_filt <- results_dreg[complete.cases(results_dreg$padj),]
# results_dreg_filt <- results_dreg_filt$padj < alphaValue
# dreg_n <- as.character(table(results_dreg_filt)["TRUE"])

####################################
          # Genes bodies
####################################

# Reads in count data and their conditions table.
load(paste0(countsDir,"PRO-",cellline,".standard10.without5prime25.featureCounts.RData"))
conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-",cellline,".csv")
conditionsTable_genes <- read.csv(conditionsTableFile, header = TRUE)

# Assigns the same ID to both the conditionTables row names and the counts column names.
colnames(counts$counts) <- conditionsTable_genes$sample
rownames(conditionsTable_genes) <- conditionsTable_genes$sample
counts_genes <- counts$counts

# Creates a DESeq2Dataset object and loads in counts.
dds_genes <- DESeqDataSetFromMatrix(countData = counts_genes, 
                              colData = conditionsTable_genes,
                              design = ~ treatment)

# Runs DESeq2 using the counts matrix.
dds_genes <- DESeq(dds_genes)

# 'Shrink' results using 'normal' mode.
resultsUnshrunken_genes <- results(dds_genes, 
                             contrast = contrast, 
                             alpha = alphaValue)

results_genes <- lfcShrink(dds_genes, 
                     contrast = contrast, 
                     res = resultsUnshrunken_genes, 
                     type = "normal")

results_genes_filt <- results_genes[complete.cases(results_genes$padj),]
results_genes_filt <- results_genes_filt$padj < alphaValue
genes_n <- as.character(table(results_genes_filt)["TRUE"])
````

````r
####################################
####################################
         # Running DESeq2
####################################
####################################

contrast <- c("treatment", "Nutlin", "DMSO")
alphaValue <- 0.05

baseDir <- paste0("/scratch/Users/dara6367/PRO-seq_interspecies-nutlin/Bidirectional-Flow/Human/")
countsDir <- paste0(baseDir,"featureCounts/")

# Loading HCT116 data.
load(paste0(countsDir,"PRO-HCT116.standard10.without5prime25.featureCounts.RData"))
conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-HCT116.csv")
conditionsTable_HCT116 <- read.csv(conditionsTableFile, header = TRUE)
colnames(conditionsTable_HCT116) <- c("bamFileName","sample","cellLine","treatment")
colnames(counts$counts) <- conditionsTable_HCT116$sample
rownames(conditionsTable_HCT116) <- conditionsTable_HCT116$sample
counts_HCT116 <- counts$counts

# Loading MCF7 data.
load(paste0(countsDir,"PRO-MCF7.standard10.without5prime25.featureCounts.RData"))
conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-MCF7.csv")
conditionsTable_MCF7 <- read.csv(conditionsTableFile, header = TRUE)
colnames(conditionsTable_MCF7) <- c("bamFileName","sample","cellLine","treatment")
colnames(counts$counts) <- conditionsTable_MCF7$sample
rownames(conditionsTable_MCF7) <- conditionsTable_MCF7$sample
counts_MCF7 <- counts$counts

# Loading MCF10A data.
load(paste0(countsDir,"PRO-MCF10A.standard10.without5prime25.featureCounts.RData"))
conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-MCF10A.csv")
conditionsTable_MCF10A <- read.csv(conditionsTableFile, header = TRUE)
colnames(conditionsTable_MCF10A) <- c("bamFileName","sample","cellLine","treatment")
colnames(counts$counts) <- conditionsTable_MCF10A$sample
rownames(conditionsTable_MCF10A) <- conditionsTable_MCF10A$sample
counts_MCF10A <- counts$counts

# Loading SJSA data.
load(paste0(countsDir,"PRO-SJSA.standard10.without5prime25.featureCounts.RData"))
conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-SJSA.csv")
conditionsTable_SJSA <- read.csv(conditionsTableFile, header = TRUE)
colnames(conditionsTable_SJSA) <- c("bamFileName","sample","cellLine","treatment")
colnames(counts$counts) <- conditionsTable_SJSA$sample
rownames(conditionsTable_SJSA) <- conditionsTable_SJSA$sample
counts_SJSA <- counts$counts

# Loading LCL data.
load(paste0(countsDir,"PRO-Human.standard10.without5prime25.featureCounts.RData"))
conditionsTableFile <- paste0(countsDir, "conditionsTable.PRO-Nutlin-Human.csv")
conditionsTable_LCL <- read.csv(conditionsTableFile, header = TRUE)
colnames(conditionsTable_LCL) <- c("bamFileName","sample","cellLine","treatment")
conditionsTable_LCL$cellLine <- rep("LCL",4)
colnames(counts$counts) <- conditionsTable_LCL$sample
rownames(conditionsTable_LCL) <- conditionsTable_LCL$sample
counts_LCL <- counts$counts

# Merging counts and conditions tables.
counts_all <- cbind(counts_HCT116, counts_MCF7, counts_MCF10A, counts_SJSA, counts_LCL)
conditionsTable_all <- rbind(conditionsTable_HCT116, conditionsTable_MCF7, conditionsTable_MCF10A, conditionsTable_SJSA, conditionsTable_LCL)

# Creates a DESeq2Dataset object and loads in counts. Formula = ~batch + condition
dds_all <- DESeqDataSetFromMatrix(countData = counts_all, 
                                  colData = conditionsTable_all,
                                  design = ~cellLine + treatment)

# Runs DESeq2 using the counts matrix.
dds_all <- DESeq(dds_all)

# 'Shrinks' results using 'normal' mode.
resultsUnshrunken_all <- results(dds_all, 
                                 contrast = contrast, 
                                 alpha = alphaValue)

results_all <- lfcShrink(dds_all, 
                         contrast = contrast, 
                         res = resultsUnshrunken_all, 
                         type = "normal")

results_all_filt <- results_all[complete.cases(results_all$padj),]
results_all_filt <- results_all_filt$padj < alphaValue
all_n <- as.character(table(results_all_filt)["TRUE"])
````

````r
####################################
        # Quality metrics
####################################

par(mfrow = c(1,3))
options(repr.plot.width = 15, repr.plot.height = 5)

# Shows the DESeq2 summary statistics.
summary(resultsUnshrunken_all)

# Plots the distribution bidirectionals rejected to determine the optimal threshold for
# calling statistically significant induced bidirectionals.
plot(metadata(resultsUnshrunken_all)$filterNumRej, 
     type = "b", 
     ylab = "number of rejections",
     xlab = "quantiles of filter",
     main = "All theta threshold")
lines(metadata(resultsUnshrunken_all)$lo.fit, col = "red")
abline(v = metadata(resultsUnshrunken_all)$filterTheta)


# Plots a histogram of the distribution of p-values for all tested bidirectionals.
hist(results_all$pvalue, main = "all p-values")

# Plots the dispersion estimates for the tested bidirectionals.
plotDispEsts(dds_all, main = "All Dispersion Estimates")
````
